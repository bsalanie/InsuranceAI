```python
def x_y(data, get_x, get_y, condition):
    
    l = [(get_x(d), get_y(d)) 
         for d in data if condition(d)]
    l.sort(key=lambda xy: xy[0])
    return list(zip(*l))

def plot_x_against_y_given_some_data(some_data, get_x, get_y_and_conditions, title, xlabel, ylabel, legends):
    
    legends_actually_found = []
    for i, (get_y, condition) in enumerate(get_y_and_conditions):
        ret = x_y(some_data, get_x, get_y, condition)
        if len(ret) == 0:
            continue
        else:
            plt.plot(ret[0], ret[1])
            legends_actually_found.append(legends[i])
    
    #plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel) 
    plt.gca().legend(legends)
    plt.gcf().set_size_inches(18.5, 10.5)
```

```python
plot_x_against_y_given_some_data(
    some_data=all_final_data,
    get_x=lambda d: d["rho_test"],
    get_y_and_conditions=[
        (lambda d: np.sum(d["T_list"][1:] < -1.64)/(len(d["T_list"])-1), lambda d: d["mu_4_choice"]=="simple"),
        (lambda d: np.sum(d["T_list"][1:] < -1.64)/(len(d["T_list"])-1), lambda d: d["mu_4_choice"]=="hard"),
        (lambda d: np.sum(d["T_list"][1:] < -1.64)/(len(d["T_list"])-1), lambda d: d["mu_4_choice"]=="none"),
    ],
    title="Fraction of T values that led to rejection of the null hypothesis",
    xlabel="Target \u03C1",
    ylabel="Fraction of rejections",
    legends=["Identity \u03A9", "Optimal \u03A9"]
)
```

```python
plot_x_against_y_given_some_data(
    some_data=all_final_data,
    get_x=lambda d: d["rho_test"],
    get_y_and_conditions=[
        (lambda d: d["negative_samples"], lambda d: d["mu_4_choice"]=="simple"),
        (lambda d: d["negative_samples"], lambda d: d["mu_4_choice"]=="hard"),
        (lambda d: d["negative_samples"], lambda d: d["mu_4_choice"]=="none"),
    ],
    title="Number of samples violating the Frechet-Hoeffding condition (deleted or modified)",
    xlabel="Target \u03C1",
    ylabel="Number of the samples",
    legends=["Identity \u03A9", "Optimal \u03A9"]
)
```

```python
plot_x_against_y_given_some_data(
    some_data=all_final_data,
    get_x=lambda d: d["rho_test"],
    get_y_and_conditions=[
        (lambda d: np.mean(d["ro_list"][1:]), lambda d: d["mu_4_choice"]=="simple"),
        (lambda d: np.mean(d["ro_list"][1:]), lambda d: d["mu_4_choice"]=="hard"),
        (lambda d: np.mean(d["ro_list"][1:]), lambda d: d["mu_4_choice"]=="none"),
    ],
    title="Target versus estimated values of \u03C1",
    xlabel="Target \u03C1",
    ylabel="Estimated \u03C1",
    legends=["Identity \u03A9", "Optimal \u03A9"]
)
```
